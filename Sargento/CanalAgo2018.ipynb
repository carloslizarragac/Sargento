```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df_dist_canal1 = pd.read_csv('sargento-canal-080918.csv', header=None, skiprows=2, 
                            names=['index', 'Date','Pres', 'Temp', 'WLevel'])
df_dist_canal2 = pd.read_csv('sargento-canal-061018.csv', header=None, skiprows=2, 
                            names=['index', 'Date','Pres', 'Temp', 'WLevel'])

df_dist_canal = pd.concat([df_dist_canal1, df_dist_canal2])

df_dist_sitio1 = pd.read_csv('sargento-old-080918.csv', header=None, skiprows=2, 
                            names=['index', 'Date','Pres', 'Temp', 'WLevel'])
df_dist_sitio2 = pd.read_csv('sargento-old-061018.csv', header=None, skiprows=2, 
                            names=['index', 'Date','Pres', 'Temp', 'WLevel'])

df_dist_sitio = pd.concat([df_dist_sitio1, df_dist_sitio2])

df_dist_salinidad = pd.read_csv('salinidad_080918.csv', header=None, skiprows=2, 
                            names=['index', 'Date','CondLow', 'CondHigh', 'Temp', 'Conduct', 
                                   'Salinity'])
df_dist_canal['Date'] =  pd.to_datetime(df_dist_canal['Date'], format='%m/%d/%Y %H:%M:%S')
df_dist_sitio['Date'] =  pd.to_datetime(df_dist_sitio['Date'], format='%m/%d/%Y %H:%M:%S')
df_dist_salinidad['Date'] =  pd.to_datetime(df_dist_salinidad['Date'], format='%m/%d/%Y %H:%M:%S')
```

```python
# Define time windows dataframes 
# One day of data
df_day_canal = df_dist_canal[(df_dist_canal['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_canal['Date'] <= '2018-08-19 00:00:00')]
df_day_sitio = df_dist_sitio[(df_dist_sitio['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_sitio['Date'] <= '2018-08-19 00:00:00')]
df_day_salinidad = df_dist_salinidad[(df_dist_salinidad['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_salinidad['Date'] <= '2018-08-19 00:00:00')]
# One week of data
df_week_canal = df_dist_canal[(df_dist_canal['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_canal['Date'] <= '2018-08-25 00:00:00')]
df_week_sitio = df_dist_sitio[(df_dist_sitio['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_sitio['Date'] <= '2018-08-25 00:00:00')]
df_week_salinidad = df_dist_salinidad[(df_dist_salinidad['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_salinidad['Date'] <= '2018-08-25 00:00:00')]

# One month of data
df_month_canal = df_dist_canal[(df_dist_canal['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_canal['Date'] <= '2018-09-19 00:00:00')]
df_month_sitio = df_dist_sitio[(df_dist_sitio['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_sitio['Date'] <= '2018-09-19 00:00:00')]
df_month_salinidad = df_dist_salinidad[(df_dist_salinidad['Date'] > '2018-08-18 00:00:00') &
                                (df_dist_salinidad['Date'] <= '2018-09-19 00:00:00')]
```

```python

 
# Load data
t = df_day_sitio['Date']
data1 = df_day_sitio['WLevel']
data2 = df_day_canal['WLevel']


# style
plt.style.use('seaborn-darkgrid')

fig, ax1 = plt.subplots()

color = 'tab:green'
ax1.set_xlabel('Date', fontsize=20)
ax1.set_ylabel('Station Water Level (m)', color=color, fontsize=16)
ax1.plot(t, data1, color=color, linewidth=3, markersize=14)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Channel Water Level (m)', color=color, fontsize=16)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color, linewidth=3, markersize=14)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped

#plt.legend(('Station', 'Channel'),loc='best')

plt.title('Daily Sea Level Variation', fontsize=20)
ax1.xaxis.set_tick_params(labelsize=16)
ax1.yaxis.set_tick_params(labelsize=16)
ax2.yaxis.set_tick_params(labelsize=16)

plt.show()
```
